#### Notebook to move SEA-AD's snrnaSeq middle temporal gyrus files into a new folder since they were erroneously placed with the dorsolateral prefrontal cortex data

```python
import synapseclient
from synapseclient import Project, File, Folder
from synapseclient import Schema, Column, Table, Row, RowSet, as_table_columns
import itertools
import pandas as pd
import numpy as np
```

```python
syn = synapseclient.Synapse()
syn.login()
```

```python
snrna_dlpdc_and_mtg_query = syn.tableQuery(f"SELECT * FROM {syn.get('syn61680783').id}")
snrna_dlpdc_and_mtg = pd.read_csv(snrna_dlpdc_and_mtg_query.filepath)
```

```python
snrna_mtg_ids = snrna_dlpdc_and_mtg.loc[snrna_dlpdc_and_mtg['tissue'] == 'middle temporal gyrus', 'id'].tolist()
snrna_mtg_ids
```

```python
for sm_id in snrna_mtg_ids:
    syn.move(sm_id, 'syn61680896')
```
